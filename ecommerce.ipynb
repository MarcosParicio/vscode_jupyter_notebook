# Crear Dataframe

```python
import pandas as pd

# Datos de ejemplo
data = {
    'IDPedido': [1, 2, 3, 4, 5],
    'IDCliente': [101, 102, 103, 104, 105],
    'Producto': ['Laptop', 'Smartphone', 'Tablet', 'Auriculares', 'Reloj Inteligente'],
    'Cantidad': [1, 2, 1, 3, 2],
    'Precio': [1000, 500, 300, 150, 200],
    'FechaPedido': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)
```

# Calcular Métricas

```python
# Calcular el total de ventas
df['Total'] = df['Cantidad'] * df['Precio']
print(df)

# Calcular el total de productos vendidos
total_productos = df['Cantidad'].sum()
print('Total productos vendidos:', total_productos)

# Calcular el promedio de ventas
promedio_ventas = df['Total'].mean()
print('Promedio de ventas:', promedio_ventas)
```

# Gráfico de Artículos Vendidos por Productos

```python
import matplotlib.pyplot as plt

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(df['Producto'], df['Cantidad'], color='skyblue')

# Añadir títulos y etiquetas
plt.title('Cantidad de Artículos Vendidos por Producto')
plt.xlabel('Producto')
plt.ylabel('Cantidad Vendida')

# Mostrar el gráfico
plt.show()
```

# Total Vendido por Producto

```python
# Crear el gráfico de barras para el total vendido por producto
plt.figure(figsize=(10, 6))
plt.bar(df['Producto'], df['Total'], color='lightgreen')

# Añadir títulos y etiquetas
plt.title('Total Vendido por Producto')
plt.xlabel('Producto')
plt.ylabel('Total Vendido')

# Mostrar el gráfico
plt.show()
```
